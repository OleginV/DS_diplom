```python
# Сред. отклонение
L = [] #пустой список
# рассчитаем среднее отклонение для каждого столбца (с пом. цикла for)
for col in list(hr.columns)[:-2]:
    line = []
    line.append(col)
    line.append(stat.pstdev(hr[col])) #создание строки для датафрейма
    L.append(line) # добавление строки в список L
# добавление данных в датафрейм
std_dev = pd.DataFrame(L, columns=["Depatment", "Std_dev"])
std_dev
```

#### Задание 3. <br>Рассчитать и визуализировать корреляционную матрицу для количественных переменных. <br>Определить две самые скоррелированные и две наименее скоррелированные переменные.

```python
# Матрица корреляций
hr.corr()
```

```python
# Визуализация матрицы корреляций с помощью heatmap
hrcorr = hr.corr()
mask = np.zeros_like(hrcorr, dtype=np.bool) # скрываем дублирующие и единичные значения матрицы
mask[np.triu_indices_from(mask)] = True

sns.heatmap(hrcorr, mask=mask, annot=True, cmap='RdYlGn', linewidths=0.2) 
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```

```python
#Две наименее и две наиболее скоррелированные переменные
min_corr = hr[['time_spend_company', 'Work_accident']].corr()
max_corr = hr[['number_project', 'average_montly_hours']].corr()
# вывод результатов расчета
print("1. Наименее скоррелированные переменные - это {} и {}, коэфф. корреляции равен: {:.4f}".format(min_corr.index[0],\
                                                                                                  min_corr.index[1],\
                                                                                                  min_corr.iloc[0,1]))
print("2. Наиболее скоррелированные переменные - это {} и {}, коэфф. корреляции равен: {:.4f}".format(max_corr.index[0],\
                                                                                                  max_corr.index[1],\
                                                                                                  max_corr.iloc[0,1]))
```

Набор данных содержит слабоскоррелированные переменные (максимальное значение коэффициента корреляции между переменными равно 0.4172) <br>

#### Задание 4. Рассчитать сколько сотрудников работает в каждом департаменте.

```python
# Численность персонала по департаментам
# создание набора данных для визуализации 
hrstaff = hr['department'].value_counts()
x = list(hrstaff.index)
y = list(hrstaff.values)

fig, ax = plt.subplots(figsize=(10,6))
width = 0.9 # ширина баров
ind = np.arange(len(y))  # расположение значений для групп
ax.barh(x, y, width) # график горизонтальных баров
ax.set_yticks(ind+width/2) # задание ширины между барами
ax.set_yticklabels(x, verticalalignment='bottom', horizontalalignment='right', fontweight='bold') # парамеитры по оси y
plt.title('Number of staff in departments', fontweight='bold') # заголовок графика
plt.xlabel("Number of staff", fontweight='bold') # заголовок оси x
plt.ylabel("Departments", fontweight='bold') # заголовок оси y
ax.invert_yaxis() # обратный порядок отображения по оси y
# подсписи значений баров 
for i, v in enumerate(y):
    ax.text(v + 3, i + .25, str(v), verticalalignment='bottom', horizontalalignment='left', fontweight='bold')
plt.show()
# таблица значений численности персонала по департаментам
print("Численность персонала в департаментах:", hrstaff, sep='\n')
```

#### Задание 6. Показать распределение сотрудников по зарплатам в каждом департаменте по отдельности.

```python
# Распределение сотрудников по уровню заработной платы по каждому департаменту
fig, ax = plt.subplots(figsize=(15,7))
sns.countplot(x="department", hue='salary', data=hr)
plt.title('Distribution salary groups on the departments', fontweight='bold') # заголовок графика
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),ha = 'center', \
                va = 'center', xytext = (0, 9), textcoords = 'offset points')
plt.show()
```

```python
# Визуализация данных о каждому департаменту в отдельности
for d in hr["department"].unique():
    plt.figure(figsize=(10,6))
    ax = sns.countplot(x=hr["department"][hr["department"]==d], hue='salary', data=hr)
    plt.title('Distribution salary groups in {}'.format(d), fontweight='bold') # заголовок графика
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),ha = 'center', \
                    va = 'center', xytext = (0, 9), textcoords = 'offset points')
    plt.show()
```

```python
# Визуализация зарплатных групп в разрезе департаментов
plt.figure(figsize=(10,6))
ax = sns.countplot(x="salary", hue='department', data=hr)
plt.title('Distribution of departments on the salary groups', fontweight='bold') # заголовок графика
for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', \
                    va = 'center', xytext = (0, 9), textcoords = 'offset points')
plt.show()
```

```python
# Создание датафрейма с переменными 'Среднее количество часов на рабочем месте в месяц'
work_time = hr[['average_montly_hours', 'salary']][hr['salary'] !='medium']
```

```python
# Визуализация данных с помощью графика распределения
plt.figure(figsize=(10,6))
sns.histplot(data=work_time, x='average_montly_hours', hue='salary')
plt.title('Distribution of time at work by salary groups', fontweight='bold');
```

```python
# Визуализация данных с помощью графика boxplot
plt.figure(figsize=(10,6))
ax = sns.boxplot(x='salary', y='average_montly_hours', data=work_time, color='#0000ff')
plt.title('Distribution of time at work by salary groups', fontweight='bold')
plt.show()
```

```python
# Оценка различий между двумя выборками с помощью Критерия Стьюдента (t-test)
t, t_pvalue = sts.ttest_ind(work_time['average_montly_hours'][work_time['salary'] =='high'], \
                           work_time['average_montly_hours'][work_time['salary'] =='low']) 
print("Значение p-value рассчитанное с помощью Критерия Стьюдента: {}".format(t_pvalue))

# Оценка различий между двумя выборками с помощью дисперсионного анализа (ANOVA)
fvalue, da_pvalue = sts.f_oneway(work_time['average_montly_hours'][work_time['salary'] =='high'], \
                              work_time['average_montly_hours'][work_time['salary'] =='low'])
print("Значение p-value рассчитанное с помощью дисперсионного анализа: {}".format(da_pvalue))

# Оценка различий между двумя выборками с помощью Критерия Манна-Уитни (U-тест)
stat, u_pvalue = sts.mannwhitneyu(work_time['average_montly_hours'][work_time['salary'] =='high'], \
                           work_time['average_montly_hours'][work_time['salary'] =='low'])
print("Значение p-value рассчитанное с помощью Критерия Манна-Уитни: {}".format(u_pvalue))
```

Таким образом, <strong><em>гипотеза о том, что сотрудники с высокими окладами проводят на работе больше времени, чем сотрудники с низкими окладами не подтверждается</em></strong>.

```python
# разделяем выборку на тренировочную и тестовую
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.25, random_state=42)
```

#### Построение модели LDA

```python
# Модель LDA
lda = LinearDiscriminantAnalysis()
# обучение данные на модели LDA
lda.fit(X_train, y_train)
```

```python
# коэффициенты дискриминатных линий
lda.coef_
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

#### Оценка качества модели на тестовой выборке

```python
# Оценка точности прогнозирования модели
acc = accuracy_score(y_test, lda.predict(X_test))
prec = precision_score(y_test, lda.predict(X_test))
recall = recall_score(y_test, lda.predict(X_test))
print("Accuracy: {:.1%}".format(acc), "Precission: {:.1%}".format(prec), "Recall: {:.1%}".format(recall), sep='\n')
```

#### Применение метода кросс-валидации для обучения и оценки качества модели LDA

```python
# Определение метода кросс-валидации модели
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=42)

# оценка модели
acc_scores = cross_val_score(lda, data, target, scoring='accuracy', cv=cv)
prec_scores = cross_val_score(lda, data, target, scoring='precision', cv=cv)
recall_scores = cross_val_score(lda, data, target, scoring='recall', cv=cv)
print("Accuracy: {:.1%}".format(np.mean(acc_scores)), "Precission: {:.1%}".format(np.mean(prec_scores)), \
      "Recall: {:.1%}".format(np.mean(recall_scores)), sep='\n')
```
